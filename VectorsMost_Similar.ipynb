### Finding the most similar vectors to our seed list terms

#### Racial and General terms

```python
terms = ['black', 'blacks', 'coloured', 'colored', 'ethnic', 'kaffir', 'kaffer', 'mulatto','mulattoes', 'native', 'natives', 'negress', 'negro', 'nigger', 'negroes', 'pickaninny', 'pygmy', 'pygmies', 'savages', 'savage', 'africans', 'african', 'locals', 'local', 'cheif', 'chiefs', 'macaca', 'macaque', 'creole', 'creoles', 'gypsies', 'gypsy']
```

```python
for i in terms:
    print('Vectors most similar to:',i)
    sort_list.append(vectors.most_similar(i))
    print(vectors.most_similar(i))
    print('\n')
```

### Sorting the list of most similar vectors by frequency:

```python
sort_list
```

```python
#flatening the sorted list for ease of data analysis:
flat_list = [item for sublist in sort_list for item in sublist]
```

```python
flat_list
```

```python
#extracting just the most similar words and not their similarity:
similar_vectors = [x[0] for x in flat_list]
```

```python
#Initializing an empty dictionary to store the frequency results:
mydict = {}
```

```python
similar_vectors
```

```python
#counting the frequency of the terms in the list:
for word in similar_vectors:
    if (word in mydict):
        mydict[word] += 1
    else:
        mydict[word] = 1
```

```python
#Printing the sorted dictionary:
import operator
sorted_dict = sorted(mydict.items(), key=operator.itemgetter(1))
```

```python
sorted_dict
```
