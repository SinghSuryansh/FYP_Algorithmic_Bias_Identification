```python
#Importing the required libraries:
from gensim.models import Word2Vec
import warnings
warnings.filterwarnings('ignore')
import operator
#Importing the libraries for generating visualisations:
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

# Loading in the word embedding model:

```python
#loading in the model:
model = Word2Vec.load("C:/Users/Suryansh/Desktop/ucd/Stage4/FYP/New/Model/Bias_Model/Bias.model")
```

```python
#Extracting the vectors and then deleting the model inorder to manage memory:
vectors = model.wv
del model
```

# Finding the most similar vectors to our seed list terms (describing the Africans):

```python
#Declaring an empty list for frequency analysis purposes:
bl_list = []
```

# Finding the most similar vectors to our seed list terms (describing Whites):

```python
#Declaring an empty list for frequency analysis purposes:
w_list = []
```

```python
brit_terms = ['whites', 'british', 'english', 'white', 'britishers', 'englishmen', 'englishman', 'european', 'europeans', 'irishmen', 'frenchmen', 'germans', 'americans']
```

```python
for d in brit_terms:
    print('Vectors most similar to ',d,':')
    w_list.append(vectors.most_similar(d))
    print(vectors.most_similar(d))
    print('\n')
```

# Frequence Analysis:

## For vectors referring to africans:

```python
#Falttening the list:
flat_bl_list = [item for sublist in bl_list for item in sublist]
#Extracting just the vector names and not their similarities:
sim_bl_vectors = [e[0] for e in flat_bl_list]
```

```python
#Initialising an empty dictionary foe storing the frequency results:
mydict = {}

#Counting the frequency of elements in the list:
for word in sim_bl_vectors:
    if (word in mydict):
        mydict[word] += 1
    else:
        mydict[word] = 1
```

```python
#Printing the sorted dictionary:
sorted_bl_dict = sorted(mydict.items(), key=operator.itemgetter(1))
#Printing the dict:
sorted_bl_dict
```
