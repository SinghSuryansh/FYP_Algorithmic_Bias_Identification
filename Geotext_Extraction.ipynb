# Geotext Extraction

```python
#Importing the required libraries:
import io, os
import csv
from geotext import GeoText
```

```python
#setting up the datapath:
path = 'D:/FYP_Data/Travelit/Travelit/'

#Empty String to store all the text together
text = ""

#Reading in the data
for root, directories, files in os.walk(path):
    for filename in files:
        filepath = os.path.join(root, filename)
        
        with open(filepath,"r",encoding="latin") as f:
            content = f.read()
            text+= content
            text+= " "
```

# Extracting the name of Geographical locations

```python
Places = GeoText(text)
```

```python
#Extracting the name of countries for our seed list:
countries = Places.countries
```

```python
#Removing the duplicates from the list:
countries = sorted(set(countries),key=countries.index)
#Printing the list of countries:
countries
```

```python
#List of countries in Africa:
africa = ['Algeria','Angola','Benin','Botswana','Burkina Faso','Burundi','Cabo Verde','Cape Verde','Cameroon','Central African Republic','Chad','Comoros','Congo','Ivory Coast','Dijbouti','Egypt','Equatorial Guinea','Eritrea','Eswatini','Swaziland','Ethiopia','Gabon','Gambia','Ghana','Guinea','Guinea-Bissau','Kenya','Lesotho','Liberia','Libya','Madagascar','Malawi','Mali','Mauritania','Mauritius','Morocco','Mozambique','Namibia','Niger','Nigeria','Rwanda','Sao Tome','Principe','Senegal','Seychelles','Sierra Leone','Somalia','South Africa','South Sudan','Sudan','Tanzania','Togo','Tunisia','Uganda','Zambia','Zimbabwe']
```

### Finding the name of African countries mentioned in the text

```python
#Converting the two lists to a set:
a = set(countries)
b = set(africa)
c = a.intersection(b)

#Convert the intersected set back into a list:
countries = list(c)
```

```python
#Saving the list into a csv file:
with open("C:/Users/Suryansh/Desktop/ucd/Stage4/FYP/Generated_Results/countries_list.csv",'w') as resultFile:
        for entries in countries:
            resultFile.write(entries)
            resultFile.write("\n")
```
